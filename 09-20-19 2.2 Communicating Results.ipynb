# Plot histograms and pdfs

```python
import numpy as np
from scipy import stats
from scipy.stats import gamma
from scipy.stats import norm
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
# Gamma

plt.figure(figsize=(8, 4)) # Create a new figure with a particular size
x = np.linspace(0, 10, 100) # Array with 100 linearly spaced points in the range [-10,10]

# Plot 4 gamma pdfs with different a (alpha) and scale (1/beta) parameters
# 1. alpha varies, beta constant
plt.plot(x, stats.gamma.pdf(x, a=2), label='alpha = 2')
plt.plot(x, stats.gamma.pdf(x, a=4), label='alpha = 4')

# 2. alpha constant, beta varies
plt.plot(x, stats.gamma.pdf(x, a=1, scale=1), label='beta = 1')
plt.plot(x, stats.gamma.pdf(x, a=1, scale=0.25), label='beta = 4')
plt.legend()
plt.title('Probability density function of a gamma distribution with varying parameters')
plt.show()

# Plot a histogram of random samples from the gamma distribution
samples = gamma.rvs(a=2, size=100)
plt.plot(x, stats.gamma.pdf(x, a=2), label='alpha = 2', linewidth=2)
plt.hist(samples, range=[0, 10], color="blue", alpha=0.2)
plt.legend()
plt.title('Histogram of random samples plotted against probability density function of a gamma distribution')
plt.show()
```

```python
# Normal

plt.figure(figsize=(8, 4)) # Create a new figure with a particular size
x = np.linspace(-10, 10, 100) # Array with 100 linearly spaced points in the range [-10,10]

# Plot 4 normal pdfs with different loc (mu) and scale (sigma) parameters
# 1. mu varies, sigma constant (default 1)
plt.plot(x, stats.norm.pdf(x, loc=-4), label='mu = -4')
plt.plot(x, stats.norm.pdf(x, loc=4), label='mu = 4')

# 2. mu constant (default 0), sigma varies
plt.plot(x, stats.norm.pdf(x, scale=2), label='sigma = 2')
plt.plot(x, stats.norm.pdf(x, scale=1/2), label='sigma = 1/2')
plt.legend() # Show the legend in the top right corner
plt.show()

# Plot a histogram of random samples from the normal distribution
samples = norm.rvs(loc=-4, size=100)
plt.plot(x, stats.norm.pdf(x, loc=-4), label='mu = -4', linewidth=2)
plt.hist(samples, range=[-10, 10], color="blue", alpha=0.2)
plt.legend()
plt.title('Histogram of random samples plotted against probability density function of a normal distribution')
plt.show()
```

```python
# Beta

plt.figure(figsize=(8, 4)) # Create a new figure with a particular size
x = np.linspace(-10, 10, 100) # Array with 100 linearly spaced points in the range [-10,10]

# Plot 4 normal pdfs with different alpha and beta parameters
# 1. alpha varies, beta constant
plt.plot(x, stats.beta.pdf(x, a=2, b=1), label='alpha = 2')
plt.plot(x, stats.beta.pdf(x, a=4, b=1), label='alpha = 4')

# 2. alpha constant, beta varies
plt.plot(x, stats.beta.pdf(x, a=1, b=1), label='beta = 1')
plt.plot(x, stats.beta.pdf(x, a=1, b=4), label='beta = 4')
plt.legend() # Show the legend in the top right corner
plt.show()

# Plot a histogram of random samples from the beta distribution
samples = beta.rvs(a=2, b=1, size=100)
plt.plot(x, stats.beta.pdf(x, a=2, b=1), label='a = 2', linewidth=2)
plt.hist(samples, range=[-10, 10], color="blue", alpha=0.2)
plt.legend()
plt.title('Histogram of random samples plotted against probability density function of a beta distribution')
plt.show()
```
